```python
import json


with open('meteorites.json', 'r') as f:
    dati_meteoriti = json.load(f)


print("Download completato!")
print("Numero totale di meteoriti caricati:", len(dati_meteoriti))


print("Chiavi del primo meteorite:")
print(dati_meteoriti[0].keys())


print("Primi 5 meteoriti:")
for i in range(5):
    print(dati_meteoriti[i], "\n")
```

```python
masse = []

for item in dati_meteoriti:
    m = item.get("mass")
    try:
        m = float(m)
        if m > 0:
            masse.append(m)
    except:
        pass

print("Masse valide trovate:", len(masse))

import matplotlib.pyplot as plt

for bins in [20, 50, 100]:
    plt.figure(figsize=(10,5))
    plt.hist(masse, bins=bins)
    plt.title(f"Distribuzione delle masse dei meteoriti ({bins} bins)")
    plt.xlabel("Massa (grammi)")
    plt.ylabel("Frequenza")
    plt.show()

plt.figure(figsize=(10,5))
plt.hist(masse, bins=50)
plt.yscale("log")
plt.title("Distribuzione delle masse dei meteoriti (scala logaritmica sull'asse Y)")
plt.xlabel("Massa (grammi)")
plt.ylabel("Frequenza (log)")
plt.show()
```
