```python
import json


with open('meteorites.json', 'r') as f:
    dati_meteoriti = json.load(f)


print("Download completato!")
print("Numero totale di meteoriti caricati:", len(dati_meteoriti))


print("Chiavi del primo meteorite:")
print(dati_meteoriti[0].keys())


print("Primi 5 meteoriti:")
for i in range(5):
    print(dati_meteoriti[i], "\n")
```

```python
masse = []

for item in dati_meteoriti:
    m = item.get("mass")
    try:
        m = float(m)
        if m > 0:
            masse.append(m)
    except:
        pass

print("Masse valide trovate:", len(masse))

import matplotlib.pyplot as plt

for bins in [20, 50, 100]:
    plt.figure(figsize=(10,5))
    plt.hist(masse, bins=bins)
    plt.title(f"Distribuzione delle masse dei meteoriti ({bins} bins)")
    plt.xlabel("Massa (grammi)")
    plt.ylabel("Frequenza")
    plt.show()

plt.figure(figsize=(10,5))
plt.hist(masse, bins=50)
plt.yscale("log")
plt.title("Distribuzione delle masse dei meteoriti (scala logaritmica sull'asse Y)")
plt.xlabel("Massa (grammi)")
plt.ylabel("Frequenza (log)")
plt.show()
```

```python
import folium
import numpy as np
import pandas as pd
```

```python
df_meteoriti = pd.DataFrame(dati_meteoriti)
df_meteoriti['lat'] = pd.to_numeric(df_meteoriti['lat'], errors='coerce')
df_meteoriti['long'] = pd.to_numeric(df_meteoriti['long'], errors='coerce')
df_meteoriti_geoloc = df_meteoriti.dropna(subset=['lat', 'long'])
media_lat = np.nanmean(df_meteoriti_geoloc['lat'])
media_lon = np.nanmean(df_meteoriti_geoloc['long'])
mappa_meteoriti = folium.Map(location=[media_lat, media_lon], zoom_start=2)
```

```python
import folium
import numpy as np
mappa_meteoriti = folium.Map(location=[0, 0], zoom_start=2)
for item in dati_meteoriti[:100]:
  if item.get("lat") and item.get("long") and item.get("name"):
    try:
      lat = float(item.get("lat"))
      lon = float(item.get("long"))
      if not np.isnan(lat) and not np.isnan(lon):
        folium.Marker([lat, lon], popup=item.get("name")).add_to(mappa_meteoriti)
    except:
      pass
mappa_meteoriti
```
